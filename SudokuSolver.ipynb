```python
import numpy as np

#checks a given list for duplicates; if there are duplicates it returns True
def check_duplicates(l):
    
    #for ignoring zeroes
    numberOfZeroes = len(l) - np.count_nonzero(l)
    
    if numberOfZeroes == 0:
        return ((len(l)) != len(np.unique(l)))
    else:
        return ((len(l) - numberOfZeroes) != (len(np.unique(l)) - 1))
    
def find_small_square(x, y, sdoku):
    squareStartX = (x//3)*3
    squareStartY = (y//3)*3
    tmpList = np.zeros(9)
    tmpIndex = 0
    for i in range(0, 3):
        for j in range(0, 3):
            tmpList[tmpIndex] = sdoku[squareStartX+i][squareStartY+j]
            tmpIndex = tmpIndex + 1
    return tmpList

def check_if_solution_found(s):
    for i in range(0,9):
        for j in range(0,9):
            if s[i][j] == 0:
                return False
    return True
```

```python
def sudoku_solve(sudoku):
    
    rows = sudoku
    #matrix transposed
    columns = sudoku.T
            
    #check if solution found
    if check_if_solution_found(sudoku):
        return True
            
    #find empty location
    for i in range(0, 9):
        for j in range(0, 9):
            if sudoku[i][j] == 0:
                currentRow = i
                currentColumn = j
                break
        if(sudoku[i][j] == 0):
            break
            
    #try 1-9
    for i in range(1,10):
        sudoku[currentRow][currentColumn] = i
        
        #check row, column, subSquares
        if not check_duplicates(rows[currentRow]) and not check_duplicates(columns[currentColumn]) and not check_duplicates(find_small_square(currentRow, currentColumn, sudoku)):
            #if i in this position has not caused invalidity, continue
            if sudoku_solve(sudoku):
                return True
            
    #if all numbers in the range 1,10 led to invadility, backtrack
    sudoku[currentRow][currentColumn] = 0
    return False
        

def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.
    """
    
    s = sudoku.copy()
    
    if sudoku_solve(s):
        return s
    else:
        for i in range(0,9):
            for j in range(0,9):
                s[i][j] = -1
        return s


test = np.array([[0., 0., 4., 0., 8., 3., 0., 0., 2.],
                 [0., 5., 1., 0., 0., 4., 3., 0., 0.],
                 [0., 0., 0., 0., 9., 6., 7., 1., 0.],
                 [1., 2., 0., 8., 0., 0., 0., 0., 6.],
                 [0., 4., 0., 0., 0., 0., 5., 0., 0.],
                 [8., 3., 0., 6., 0., 7., 9., 0., 0.],
                 [0., 6., 0., 3., 0., 9., 0., 4., 0.],
                 [0., 0., 7., 0., 0., 0., 2., 0., 5.],
                 [0., 9., 0., 0., 5., 0., 8., 0., 3.]])

print("Sudoku Puzzle: \n", test)
print("Solution: \n", sudoku_solver(test))
```
